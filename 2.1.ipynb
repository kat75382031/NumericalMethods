```python
#example
A=[[20.,4.,-8.],
  [-3.,15.,5.],
  [6.,3.,-18.]]
b=[1.,-2.,3.]
x=[0.,0.,0.]
x0=[0.]*3
x_0=[0.]*3
xz=[0.]*3
xz0=[0.]*3
xz_0=[0.]*3

def Bbeta(A,b):
    for i in range(0,3):
        sum = 0
        for j in range (0,2):
            if j!=i: 
                sum+=A[i][j]
        if (A[i][i]>0 and A[i][i]>sum) or (A[i][i]<0 and A[i][i]<sum):
            b[i]/=A[i][i]
            for k in reversed(range(3)):
                if k!=i:
                    A[i][k]/=-A[i][i]
            A[i][i]=0
            
def normA1(A):
    max=0
    for i in range(0,3):
        sum=0
        for j in range(0,3):
            sum+=abs(A[i][j])
        if sum>max:
            max=sum
    return max

def normA2(A):
    max=0
    for i in range(0,3):
        sum=0
        for j in range(0,3):
            sum+=abs(A[j][i])
        if sum>max:
            max=sum
    return max

def normb1(b):
    max=0
    for i in range(0,3):
        if abs(b[i])>max:
            max=abs(b[i])
    return max

def normb2(b):
    return abs(b[0])+abs(b[1])+abs(b[2])

def accuracy1(i,l,N,n):
    eps=1
    for i in range(0,i):
        eps*=N
    eps*=n
    eps/=1-N
    return eps

def accuracy2(x_1,x_0):
    d=[0.,0.,0.]
    x1=[0.,0.,0.]
    x2=[0.,0.,0.]
    for i in range(0,3):
        x1[i]=x_0[i]
        x2[i]=x_1[i]
    for j in range(0,3):
        d[j]=x1[j]-x2[j]
    if l==1:
        delta=normb1(d)
    else:
        delta=normb2(d)
    return delta

Bbeta(A,b)
if normA1(A)<normA2(A):
    N=normA1(A)
    n=normb1(b)
    l=1
    if normA1(A)<1:
        print('Итерационный процесс сходится, т.к. норма матрицы В=%.10f<1' % N)
if normA2(A)<normA1(A):
    N=normA2(A)
    n=normb2(b)
    l=2
    if normA2(A)<1:
        print('Итерационный процесс сходится,т.к. норма матрицы В=<1')   
for m in range(0,3):
    x0[m]=b[m]
for m in range(0,3):
    x_0[m]=b[m]
print("Метод простых итераций:")
print("i       x           y              z            eps          delta")
print("%.f %.10f %.10f %.10f" % (1,x0[0], x0[1], x0[2]))
for i in range(2,9):
    for j in range(0,3):
        x[j]=b[j]
        for k in range(0,3):
            x[j]+=A[j][k]*x0[k]
    print("%.f %.10f %.10f %.10f %.10f %.10f " % (i,x[0],x[1],x[2], accuracy1(i-1,l,N,n),accuracy2(x,x0)))
    for m in range(0,3):
        x0[m]=x[m]

print("")
print("Метод Зейделя:")        
B=[[20.,4.,-8.],
   [-3.,15.,5.],
   [6.,3.,-18.]] 
c=[1.,-2.,3.]
Bbeta(B,c)
for m in range(0,3):
    xz0[m]=c[m]
for m in range(0,3):
    xz_0[m]=c[m]
print("i       x           y              z            eps          delta")
print("%.f %.10f %.10f %.10f" % (0,xz0[0], xz0[1], xz0[2]))
for i in range(1,6):
    for m in range(0,3):
        xz[m]=c[m]  
    for j in range(0,3):
        xz[0]+=B[0][j]*xz0[j]
    xz[1]+=B[1][0]*xz[0]+B[1][1]*xz0[1]+B[1][2]*xz0[2]
    xz[2]+=B[2][0]*xz[0]+B[2][1]*xz[1]+B[2][2]*xz0[2]
    print("%.f %.10f %.10f %.10f %.10f %.10f " % (i,xz[0],xz[1],xz[2], accuracy1(i,l,N,n),accuracy2(xz,xz0)))
    for m in range(0,3):
        xz0[m]=xz[m]
```
