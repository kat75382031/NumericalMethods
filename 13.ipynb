```python
#example
from math import sqrt
n=5
a1=1.
b1=10.
h=(b1-a1)/(n-1)

x=[0.]*n
y=[0.]*n

a=[0.]*(n-1)
b=[0.]*n
c=[0.]*(n-1)
d=[0.]*n

def progon(a,b,c,d):
    n=len(b)-1
    y=[0]*len(b)
    x=[0]*len(b)
    alpha=[0]*n
    beta=[0]*len(b)
    y[0]=b[0]
    alpha[0]=-c[0]/y[0]
    beta[0]=d[0]/y[0]
    
    
    for i in range(1,n):
        y[i]=b[i]+alpha[i-1]*a[i-1] 
        alpha[i]=-c[i]/y[i]
        beta[i]=(d[i]-a[i-1]*beta[i-1])/y[i]
        
    y[n]=b[n]+a[n-1]*alpha[n-1]
    beta[n]=(d[n]-a[n-1]*beta[n-1])/y[n]
                                      
    for i in range(0,n+1):
        if i!=0:
            x[n-i]=alpha[n-i]*x[n-i+1]+beta[n-i]
        else:
            x[n-i]=beta[n-i]
    return x

def f(x):
    return 2*sqrt(x)

def K(x):
    return 2*x*x

def L(x):
    return x

def M(x):
    return 1

def T(x):
    return 2

def Z(x):
    return 2*sqrt(10)

def R(x):
    return 0

def S(x):
    return 1

def V(x):
    return 0

def W(x):
    return 1


for j in range(0,n):
    x[j]=a1+h*j
    if (j!=0) and (j!=n-1):
        a[j-1]=K(x[j])/h/h-L(x[j])/2/h
        b[j]=-2*K(x[j])/h/h+M(x[j])
        c[j]=K(x[j])/h/h+L(x[j])/2/h
        d[j]=f(x[j])
    else:
        if j==0:
            d[j]=T(x[j])
            b[j]=-R(x[j])/h+S(x[j])
            c[j]=R(x[j])/h
        else:
            d[j]=-Z(x[j])
            a[j-1]=V(x[j])/h
            b[j]=-V(x[j])/h-W(x[j])
print(" ")
Y=y=progon(a,b,c,d)
print("    x             y")
for j in range(0,n):
    print("%.10f %.10f" % (x[j],y[j]))
```
