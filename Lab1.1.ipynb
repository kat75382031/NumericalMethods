```python
import numpy as np
import matplotlib.pyplot as plt
# t :[0;2], x:[0,pi]
n=11
h=np.pi/(n-1)
tau=0.02
ht=100
a=1
b=1
eps=[0]*ht
x=np.linspace(0,np.pi,n)
def f1(a,b,t):
    a=-np.exp(-a*t)*(np.cos(b*t)+np.sin(b*t))
    return a
def analit(x,t):
    return np.exp(-t*a)*np.cos(x+b*t)

y=[0]*n
y0=[0]*n
# 0 
for j in range (0,n):
        y0[j]=np.cos(j*h)
#print y0
x=np.linspace(0,np.pi,n)
for i in range(0,ht):
    t=(i+1)*tau
    for j in range(1,n-1):
        y[j]=(a*tau/h/h-b*tau/2/h)*y0[j-1]+(1-2*a*tau/h/h)*y0[j]+(a*tau/h/h+b*tau/2/h)*y0[j+1]
    y[0]=-h*f1(a,b,t)/(1+h)+y[1]/(1+h)
    y[n-1]=-h*f1(a,b,t)/(1-h)+y[n-2]/(1-h)
      
    #y[0]=(2*h*f1(a,b,t)-4*y[1]+y[2])/(-2*h-3)
    #y[n-1]=(-2*h*f1(a,b,t)+4*y[n-2]-y[n-3])/(-2*h+3)
    for j in range(len(y)):
        if eps[i]<abs(analit(x[j],t)-y[j]):
            eps[i]=abs(analit(x[j],t)-y[j])
    for k in range(0,n):
        y0[k]=y[k]
    print y
print t  
```
