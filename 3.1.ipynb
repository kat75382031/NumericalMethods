```python
#example
from math import sqrt
A=[[-12.,4.,8.],
  [4.,11.,-6.],
  [8.,-6.,2.]]
#w0=[1/3**0.5,1/3**0.5,1/3**0.5]
w0=[1/sqrt(3),1/sqrt(3),1/sqrt(3)]
nu=[0.,0.,0.]
ro=[0.,0.,0.]
w=[0.,0.,0.]

def multiplication(A,b):
    m=[0.,0.,0.]
    for i in range(0,3):
        for j in range(0,3):
            m[i]+=A[i][j]*b[j]
    return m

def scal(a,b):
    return a[1]*b[1]+a[2]*b[2]+a[0]*b[0]

def norm(a):
    return (a[1]*a[1]+a[2]*a[2]+a[0]*a[0])**0.5

for m in range(1,11):
    print("i=%.d" % m)
    nu=multiplication(A,w0)
    ro=scal(nu,w0)
    n=norm(nu)
    print("Норма вектора ню: %.10f" %n)
    for i in range(0,3):
        w[i]=nu[i]/n
    print("     ro                          omega")
    print("%.10f %.10f %.10f %.10f" % (ro, w[0],w[1],w[2]))
    print(" ")
    w0=w
```
